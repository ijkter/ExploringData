## 分析背景

### 数据集意义

根据超市历史订单数据，能够反映超市运营状况，及时发现规律与问题，以指导超市运营。

### 数据集来源

本次分析所使用的数据集来源于 Kaggle 上的公开数据集 [《US Superstore data》](https://www.kaggle.com/juhi1994/superstore)。

### 数据理解

数据文件 US_Superstore_data.csv 中各项字段及其说明如下：

- Row ID：行 id
- Order ID：订单 id
- Order Date：订单日期
- Ship Date：发货日期
- Ship Mode：发货标准
- Customer ID：客户 id
- Customer Name：客户名
- Segment：客户分类
- Country：国家
- City：城市
- State：州
- Postal Code：邮政编码
- Region：地区
- Product ID：商品 id
- Category：商品分类
- Sub-Category：商品子分类
- Product Name：商品名
- Sales：价格
- Quantity：数量
- Discount：折扣
- Profit：利润

## 分析目的

本次分析的目的是：根据 2014~2017 年总共四年的订单数据，了解销售情况，并发掘有价值的信息和销售异常。主要围绕以下几个方向展开分析：

1. 整体销售情况
   - 销售和盈利
   - 订单和销量
   - 客户
   - 从时间维度分析以上销售指标变化，寻找趋势或规律
2. 商品销售情况
   - 各分类商品数量、销售额、利润、毛利率
   - 销售额高的商品，利润最高、最低的商品
3. 区域销售情况
   - 各区销售额、利润、毛利率
   - 销售情况较好、较差的区域
4. 客户情况
  - 客户类型
  - 消费情况
  - 用户分层
5. 发货情况
  - 发货标准分布
  - 发货标准与发货天数的关系

## 重要结论

1. 整体销售情况

   - 整体上，订单、销量、销售额与利润都在逐年增长，而毛利率每年稳定在 13% 左右。

   - 每年的季度销售指标第一季度当年最低，然后逐季度递增，至第四季度达当年最高。
  
   - 新客户增长逐年减缓，老用户流失逐年增加，需要拉新留旧。
  
   - 客户复购率高达 98.5%。

   - 2017 年四季度销售额、订单、销量与 2016 同期相比有大幅增长，但利润反降，推测是由于当季折扣力度过大。建议深入了解用户消费需求，作出合适的定价与折扣。
  
2. 商品销售情况

   - 商品有三种分类，其中，办公用品占 60%，家具和技术接近，分别占 20% 上下。

   - 相比办公用品和技术的毛利率水平（均为 17%），家具类商品毛利率极低（3%），平均折扣率：家具类 > 办公用品类 > 技术类。建议及时调整销售策略、根据消费者需求把控好各类商品的库存、降低折扣损失。

   - 从 2014 到 2017，各类商品销售额都在逐年增加，办公用品类和技术类商品的利润随之增长，且涨幅分别近 75% 和 110%，但家具类商品的利润一直在较低的的区间（3000~7000）内波动，且历年折扣占比明显高于其他类商品。


3. 区域销售情况

   - 总体上，各地区销售情况：西部 > 东部 > 南部 > 中部。
   
   - 西部地区和东部地区的各项销售指标都在逐年增长，毛利率稳定在 10% 以上，且西部地区毛利率始终稍高于东部区。
  
   - 南部地区和中部地区的各项销售增长不稳定，且于 2017 年毛利率有明显骤降，需要研究客户消费情况、调整销售策略。
   
   - 各地区都存在 2~3 个负盈利的州，可以向同地区盈利状况较好的州借鉴运营经验、调整销售策略。
   
   - 销售额与利润明显高出其他城市的城市依次有：纽约、洛杉矶、西雅图、旧金山。


4. 客户情况

  - 客户类型：客户类型占比由高到低依次为 'Consumer','Corporate'、 'Home Office'，贡献销售额与利润也依次递减。
  
  - 消费情况：各类型客户销售额与利润也逐年上升，贡献占比相对稳定。
  
  - 用户分层：根据 RFM 模型划分为四类用户 -- VIP 用户、高价值用户、潜在价值用户、低价值用户。


5. 发货情况

   - 发货标准在订单中的占比由低到高依次为 'Same Day'、'First Class'、 'Second Class'、'Standard Class'，而发货效率依次递减。

## 数据分析

### 整体销售情况

#### 销售与盈利

````python
total_sales = round(data['Sales'].sum(),  4)
total_profit = round(data['Profit'].sum(),  4)
total_rate = round(float(total_profit) / total_sales,  4)

table = (
    Table()
    .add(
        ['总销售额', '总利润', '毛利率'],
        [[
            total_sales, 
            total_profit, 
            total_rate
        ]]
    )
)
table.render_notebook()
````

#### 订单与销量

````python
order_des = data['OId'].describe()
sales_volume = data['Quantity'].sum()

table = (
    Table()
    .add(
        ['总订单数', '平均订单价', '总销量', '连带率'],
        [[
            order_des['unique'],
            round(total_profit / order_des['unique'], 2), 
            sales_volume,
            round(sales_volume / order_des['unique'], 2)
        ]]
    )
)
table.render_notebook()
````

查看单笔订单商品数的分布：

````python
vc = data['OId'].value_counts().value_counts()

vc_bar = (
    Bar()
    .add_xaxis(vc.index.astype(str).tolist())
    .add_yaxis('', vc.values.tolist())
)
vc_bar.render_notebook()

vc_pie = (
    Pie()
    .add(
        '单笔订单商品数',
        [z for z in zip(vc.index.astype(str), vc.values.tolist())],
        label_opts=opts.LabelOpts(formatter='{b}项: {d}%'),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="单笔订单商品数分布"),
        legend_opts=opts.LegendOpts(is_show=False)
    )
)
vc_pie.render_notebook()
````

结论：

1. 超过 90% 的订单商品数小于等于 4 项。


2. 50% 的订单只有 1 项商品。


3. 单笔订单最大商品数为 14 项。

#### 客户

````python
user_n = data['Customer ID'].nunique()

table = (
    Table()
    .add(
        ['客户数', '人均订单数', '客单价', 'ARPU'],
        [[
            user_n, 
            round(order_des['unique'] /  user_n, 4),
            round(total_sales / user_n, 4), 
            round(total_profit /  user_n, 4)
        ]]
    )
)
table.render_notebook()
````

客户购买次数分布：

````python
c_oc = data[['Customer Name', 'OId']].pivot_table(index='Customer Name',
                                    values='OId',
                                    aggfunc=lambda x: x.nunique()
                                    )
c_oc_vc = pd.Series(c_oc.values[:, 0]).value_counts()

pie = Pie()
pie.add(
    '购买次数',
    [z for z in zip(c_oc_vc.index.astype(str), c_oc_vc.values.tolist())],
    label_opts=opts.LabelOpts(formatter='{b}次:{d}%'),
)
pie.set_global_opts(
    title_opts=opts.TitleOpts(title="客户购买次数分布"),
    legend_opts=opts.LegendOpts(is_show=False)
)
pie.render_notebook()
````

结论：

客户复购率很高，达 95.5%。

每年客户的新增和流失情况：

```
注：新增客户、流失客户、净增加客户定义如下
- 新增客户：客户在某一年第一次有订单数据，客户即当年新增客户。由于订单数据时间区间为 2014~2017 年，为了避免存量用户带来的误差，不计入 2014 年新增客户。
- 流失客户：客户在某一年后再没有订单数据，客户即次年流失用户。
- 净增加客户：新增客户 - 流失客户
```

````python
# 除去 2014 年，客户最小订单所在年份即成为新用户的年份
c_fd = data[['Customer Name', 'Order Date']].groupby('Customer Name').min()
c_fd['year'] = c_fd['Order Date'].dt.year
c_new = c_fd.groupby('year').count()

# 除去 2017 年，用户流失的年份即最后一次购买年份的下一年
c_ld = data[['Customer Name', 'Order Date']].groupby('Customer Name').max()
c_ld['year'] = c_ld['Order Date'].dt.year + 1
c_lost = c_ld.groupby('year').count()

c_new_lost = pd.concat([c_new, c_lost], axis=1)
c_new_lost.columns = ['新增客户', '流失客户']
c_new_lost.dropna(inplace=True)
c_new_lost['净增加客户'] = [i -v for i, v in c_new_lost.values]

bar = (
    Bar()
    .add_xaxis(c_new_lost.index.tolist())
    .add_yaxis(
        '新增客户', 
        c_new_lost.values[:, 0].tolist()
    )
    .add_yaxis(
        '流失客户', 
        c_new_lost.values[:, 1].tolist()
    )
    .add_yaxis(
        '净增加客户', 
        c_new_lost.values[:, 2].tolist()
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='每年客户新增及流失情况')
    )
)
bar.render_notebook()
````

结论：

1. 2014 ~ 2017 年每年的客户数量整体呈增加趋势，只有 2015 年较 2014 年有小幅度（4%）下滑。



2. 每年新增用户的数量越来越少。
   
   建议：调整拉新策略。


3. 每年流失的老用户越来越多，且流失速度逐年加快，甚至在 2017 年的客户净增已经出现了负增长

   建议：除自然流失，分析客户快速流失的原因，并采取挽留措施。
   
   具体建议措施:
   - 对客户进行问卷调查，寻找流失原因；

### RFM 用户分层

本次用户分析使用 RFM 模型构建用户分类，因此需要准备的指标有：用户名、订单数、购买数、销售额、利润、客户类型、地区、第一次购买、最近一次购买。

````python
# 提取 RFM 数据：
# 1.计算订单数
c_oc = data[['Customer Name', 'OId']].pivot_table(index='Customer Name',
                                    values='OId',
                                    aggfunc=lambda x: x.nunique()
                                    )

# 2. 计算销售额
c_s = data[['Customer Name', 'Sales']].groupby(by=['Customer Name']).sum()

# 计算第一次购买日期和最近一次购买日期
c_fd = data[['Customer Name', 'Order Date']].groupby('Customer Name').min()
c_ld = data[['Customer Name', 'Order Date']].groupby('Customer Name').max()

# 3. 合并结果，修改列名
users = pd.concat([c_oc, c_s, c_fd, c_ld], axis=1)
users.columns = ['订单数', '销售额', '第一次购买日期', '最近一次购买日期']

# 4. 以 '2018-01-01' 为当前日期，计算用户“用户最近一次消费时间间隔”并只取天数
users['最近一次消费时间间隔'] = pd.to_datetime('2018-01-01') - users['最近一次购买日期']
users['最近一次消费时间间隔'] = users['最近一次消费时间间隔'].apply(lambda x: x.days)


# 计算 rfm 得分：
# 1. 数据归一化：将 R、F、M 得分归一化
r_score = 1 - (users['最近一次消费时间间隔'] - users['最近一次消费时间间隔'].min()) \
            / (users['最近一次消费时间间隔'].max() - users['最近一次消费时间间隔'].min())
f_score = (users['订单数'] - users['订单数'].min()) \
            / (users['订单数'].max() - users['订单数'].min())
m_score = (users['销售额'] - users['销售额'].min()) \
            / (users['销售额'].max() - users['销售额'].min())
rfm = pd.DataFrame({'R':r_score, 'F':f_score, 'M':m_score}, index=users.index, dtype=np.float64)

# 2. 权重计算：使用变异系数法 $w_i = \frac{\sigma}{v}$ ，计算 R、F、M 的权重
v = [
    users['最近一次消费时间间隔'].std() / users['最近一次消费时间间隔'].mean(),
    users['订单数'].std() / users['订单数'].mean(),
    users['销售额'].std() / users['销售额'].mean(),
]
w = np.array(v) / sum(v)

# 3. 计算 rfm 得分：根据权重进行加权得分计算
rfm['rfm_score'] = rfm['R'] * w[0] + rfm['F'] * w[1] + rfm['M'] * w[2]

# 4. 用户分层：根据 rfm 得分，按区间 [0, 0.5, 0.8, 0.95, 1] 划分 '低价值', '潜在价值', '高价值', 'VIP' 四类用户
rfm_labels = ['低价值', '潜在价值', '高价值', 'VIP']
rfm['label'] = pd.cut(
    rfm['rfm_score'], 
    rfm['rfm_score'].quantile([0, 0.5, 0.8, 0.95, 1]), 
    labels=rfm_labels
)

# 5. 效果检验
vip = [rfm[rfm['label'] == 'VIP'].mean().tolist()[:3]]
gjz = [rfm[rfm['label'] == '高价值'].mean().tolist()[:3]]
qzjz = [rfm[rfm['label'] == '潜在价值'].mean().tolist()[:3]]
djz = [rfm[rfm['label'] == '低价值'].mean().tolist()[:3]]

radar = (
    Radar()
    .add_schema(
        schema=[
            opts.RadarIndicatorItem(name="R", max_=1),
            opts.RadarIndicatorItem(name="F", max_=1),
            opts.RadarIndicatorItem(name="M", max_=1),
        ]
    )
    .add('VIP', vip, color='#A1232B')
    .add('高价值', gjz, color='#FA842B')
    .add('潜在价值', qzjz, color='#384C70')
    .add('低价值', djz, color='#327662')
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(title_opts=opts.TitleOpts(title='各类用户 RFM 得分均值雷达图'))
)
radar.render_notebook()
````

结论：

1. 重要价值用户是最优质的用户群体，应重点关注，既要保持其粘性，又要继续引导消费，可为这类用户提供vip服务；
2. 重要发展用户的特点是近期有消费但频次不高，策略是提高其消费次数，具体措施有促销活动提醒和优惠卷活动等；
3. 重要保持用户的特点是消费频次高但有一段时间没有消费，策略是重新唤醒，通过app消息推送，以及站外广告营销吸引其注意力，促进复购；
4. 重要挽留用户近期没有消费且频次不高，若不加以挽留，会有流失的可能，对于这类用户一方面需要保持曝光量，持续推送活动和优惠信息，另一方面需要进一步研究其兴趣和需求，
